Checking Normal Distribution using Q-Q plot

```python
smf.qqplot(data['Laboratory 1'], line='q')
plt.title("Laboratory 1")
smf.qqplot(data['Laboratory 2'], line='q')
plt.title("Laboratory 2")
smf.qqplot(data['Laboratory 3'], line='q')
plt.title("Laboratory 3")
smf.qqplot(data['Laboratory 4'], line='q')
plt.title("Laboratory 4")
plt.show()
```

Hypothesis Using T-Statistics

```python
statistic,p_value=stats.f_oneway(data.iloc[:,0],data.iloc[:,1],data.iloc[:,2],data.iloc[:,3])
print("p_value: ",p_value)
```

```python
significance_level=0.05
if p_value<=significance_level:
  print("We reject Null Hypothesis there is a difference between report of the laboratory")
else:
  print("We fail to reject Null Hypothesis")
```
